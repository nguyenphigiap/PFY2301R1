### Thực hành:
Viết chương trình nhập vào một số nguyên và cho biết số nguyên đó là số chẵn hay số lẻ.

```python
so_nguyen = int(input("nhập vào một số nguyên: "))
if so_nguyen % 2 == 0:
    print(f"số {so_nguyen} là số chẵn")
else:
    print(f"số {so_nguyen} là số lẻ")
```

# Mức độ khen thưởng như sau:
# - Nếu là nhân viên level 1 thì thưởng 1000
# - Nếu là nhân viên level 2 thì thưởng 2000

```python
total_hour = int(input("Nhập tổng số giờ làm: "))
level = int(input("Nhập level của nhân viên: "))
if total_hour > 800:
    # khen thưởng
    if level == 1:
        print("Bạn được thưởng 1000usd")
    else:
        print("bạn được nhận thưởng 2000usd")
else:
    print("Bạn chưa đạt được tiêu chí nhận thưởng!")
```

## Câu lệnh if…elif…else

```python
# Viết chương trình nhập vào số ngày trong tuần (2, 3, 4, 5, 6, 7, 8) 
# và in ra từ tiếng Anh tương ứng với ngày đó.
week_day = int(input("nhập vào số ngày: "))
if week_day == 2:
    print("Monday")
elif week_day == 3:
    print("Tuesday")
elif week_day == 4:
    print("wednesday")
elif week_day == 5:
    print("thursday")
elif week_day == 6:
    print("friday")
elif week_day == 7:
    print("saturday")
else:
    print("sunday")
```
