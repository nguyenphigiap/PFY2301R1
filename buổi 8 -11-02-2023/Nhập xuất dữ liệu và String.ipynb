## Cách nhập xuất dữ liệu

### Xuất dữ liệu ra màn hình
Để in các nội dung mà chúng ta mong muốn ra màn hình, ta sử dụng lệnh print()

```python
# Ví dụ: In ra nội dung "python is very easy"
print("python is very easy")
```

### Các cách custom với lệnh print

Để in một nội dung kèm theo giá trị của biến, ta có các cách sau:

```python
food_1 = "cookies"
food_2 = "sandwich"

# i love chocolate and I hate sandwich
# "i love " --food_1-- "and I hate" --food_2--
```

```python
# In trực tiếp
print("i love chocolote and I hate sandwich")
```

```python
# In bằng cách ghép các nội dung với nhau
print("I love" , food_1 , "and I hate " , food_2, sep="+") #sep là: separate
print("I love" , food_1 , "and I hate " , food_2, sep="+") 
print("I love " + food_1 + " and I hate " + food_2)
```

```python
# Sử dụng String Format
print(f"I love {food_1} and I hate {food_2}")
```

```python
# Sử dụng nhiều lệnh print
# \n\n  ký hiệu: \n là xuống dòng thứ 2
print("I love " , end="")
print(food_1 , end="")
print(" and I hate " , end="")
print(food_2)
```

#### Thực hành: Viết chương trình nhập vào số đo cạnh và mã màu của tam giác đều, sau đó sử dụng turtle để vẽ ra hình tam giác có số đo và màu tương ứng

```python
#  begin_fill()
#  end_fill()
canh_tam_giac_deu = int(input("nhập vào 1 cạnh tam giác đều: "))
mau_canh_tam_giac_deu = str(input("nhập mã màu cạnh tam giác đều: "))

import turtle
pen = turtle.Turtle()

pen.penup()
pen.goto(50, 0)
pen.pendown()
pen.circle(canh_tam_giac_deu , )

turtle.done()
```

```python
# chữa bài
import turtle
side = int(input("Please input side of triangle: "))
color = input("please input triangle color: ")

pen = turtle.Turtle()
pen.fillcolor(color)
pen.begin_fill()
pen.circle(side , steps = 3)
pen.end_fill()
turtle.done()
```

### Các thao tác phổ biến với string

```python
# Kiểm tra độ dài string
coure_name = "Python fudamental"
print(len(coure_name))
```

```python
# Truy xuất ký tự bằng cơ chế indexing và reverse indexing
# indexing: đánh dấu từ trái qua phải: 0 1 2 3 .....
# reverse indexing: đánh dấu từ phải qua trái :  -1 -2 -3 -4...
content = "hello"

# indexing
print(content[4])



# reverse indexing
print(content[-1])
```
