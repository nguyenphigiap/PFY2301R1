# nâng cấp ngôi nhà mơ ước

```python
import turtle

pen = turtle.Turtle()

# thiết lập nét vẽ

pen.pencolor("blue")
pen.pensize(2)
pen.speed(20)
# vẽ ngôi nhà trước
# vẽ khung nhà hình chữ nhật bên ngoài
pen.penup()
pen.goto(250, 0)
pen.pendown()
pen.right(90)
pen.forward(200)
pen.right(90)
pen.forward(400)
pen.right(90)
pen.forward(200)
pen.right(90)
pen.forward(400)

#vẽ khung cửa ra vào
pen.penup()
pen.goto(0 ,-50)
pen.pendown()
pen.forward(100)
pen.right(90)
pen.forward(150)
pen.right(90)
pen.forward(100)
pen.right(90)
pen.forward(150)
pen.right(90)
#vẽ mái nhà
pen.penup()
pen.goto(150 , 0)
pen.pendown()
pen.left(135)
pen.forward(212.132)
pen.left(90)
pen.forward(212.132)
pen.left(135)
pen.forward(300)

# vẽ ống khói
pen.penup()
pen.goto(75 , 75)
pen.pendown()
pen.left(90)
pen.forward(60)
pen.left(90)
pen.forward(33)
pen.left(90)
pen.forward(26)
pen.left(90)

# vẽ ống khói
pen.penup()
pen.goto(75, 150)
pen.pendown()
pen.left(90)
pen.forward(60)

pen.penup()
pen.goto(50 , 160)
pen.pendown()
pen.forward(40)
pen.right(90)

# Vẽ cây to bên phải

 # vẽ thân cây to và đổ màu 

pen.penup()
pen.goto(400, -300)
pen.pendown()
pen.fillcolor("grey")
pen.begin_fill()
pen.forward(100)
pen.left(100)
pen.forward(200)
pen.left(80)
pen.forward(60)
pen.left(80)
pen.forward(200)
pen.left(100)
pen.forward(50)
pen.end_fill()

# vẽ tán cây 3 tầng cây to và đổ màu
pen.penup()
pen.goto(435 , -102)
pen.pendown()
pen.color("dark green")
pen.fillcolor("dark green")
pen.begin_fill()
pen.forward(102)
pen.left(135)
pen.forward(144.25)
pen.left(90)
pen.forward(144.25)
pen.left(135)
pen.forward(102)
pen.end_fill()

pen.penup()
pen.goto(435 , 0)
pen.pendown()
pen.color("green")
pen.fillcolor("green")
pen.begin_fill()
pen.forward(80)
pen.left(135)
pen.forward(113.127)
pen.left(90)
pen.forward(113.137)
pen.left(135)
pen.forward(80)
pen.end_fill()

pen.penup()
pen.goto(435 , 80)
pen.pendown()
pen.color("light green")
pen.fillcolor("light green")
pen.begin_fill()
pen.forward(60)
pen.left(135)
pen.forward(84.853)
pen.left(90)
pen.forward(84.853)
pen.left(135)
pen.forward(60)
pen.end_fill()

# Vẽ cây bé
# vẽ thân cây bé
pen.penup()
pen.goto(-250 , -320)
pen.pendown()
pen.color("blue")
pen.fillcolor("gray")
pen.begin_fill()
pen.forward(50)
pen.left(100)
pen.forward(100)
pen.left(80)
pen.forward(35)
pen.left(80)
pen.forward(100)
pen.left(100)
pen.forward(40)
pen.end_fill()

# vẽ tán cây bé và đổ màu

pen.penup()
pen.goto(-235 , -230)
pen.pendown()
pen.color("dark green")
pen.fillcolor("dark green")
pen.begin_fill()
pen.forward(60)
pen.left(135)
pen.forward(84.853)
pen.left(90)
pen.forward(84.85)
pen.left(135)
pen.forward(60)
pen.end_fill()

pen.penup()
pen.goto(-235, -170)
pen.pendown()
pen.color("green")
pen.fillcolor("green")
pen.begin_fill()
pen.forward(50)
pen.left(135)
pen.forward(70.71)
pen.left(90)
pen.forward(70.71)
pen.left(135)
pen.forward(50)
pen.end_fill()

pen.penup()
pen.goto(-235 , -120)
pen.pendown()
pen.color("light green")
pen.fillcolor("light green")
pen.begin_fill()
pen.forward(40)
pen.left(135)
pen.forward(56.57)
pen.left(90)
pen.forward(56.57)
pen.left(135)
pen.forward(40)
pen.end_fill()

turtle.done()
```
