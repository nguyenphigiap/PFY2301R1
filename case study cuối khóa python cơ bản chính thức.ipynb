```python
class Quanlynhansu:
    def __init__(self, hoten, quequan, maphongban, luonghientai, ngayvaocongty):
        self.hoten = hoten
        self.quequan = quequan
        self.maphongban = maphongban
        self.luonghientai = luonghientai
        self.ngayvaocongty = ngayvaocongty
        

from tkinter import *
#from tkinter import messagebox
from tkinter.messagebox import showinfo

window = Tk()
window.title("Quan Ly Nhan Su")
window.geometry("500x300")  # kích thước chiều dài và chiều rộng cửa sổ

# Label
intro_lbl = Label(window, text="Thong Tin Nhan Su")
hoten_lbl = Label(window, text="Ho Ten")
quequan_lbl = Label(window, text="Que Quan")
maphongban_lbl = Label(window, text="Ma Phong Ban")
luonghientai_lbl = Label(window, text="Luong Hien Tai")
ngayvaocongty_lbl = Label(window, text="Ngay Vao Cong Ty")

timkiem_lbl = Label(window, text="Tim Kiem:")

# Entry
hoten = Entry(window)
quequan = Entry(window)
maphongban = Entry(window)
luonghientai = Entry(window)
ngayvaocongty = Entry(window)
timkiem_ten = Entry(window)

# list_patient được sử dụng để lưu trữ thông tin các bệnh nhân
list_quanlynhansu = []

def them_thongtinnhansu():
    # lấy thông tin từ các entry
    hoten_value = hoten.get()
    quequan_value = quequan.get()
    maphongban_value = maphongban.get()
    luonghientai_value = luonghientai.get()
    ngayvaocongty_value = ngayvaocongty.get()

    
    # khởi tạo đối tượng
    quanlynhansu = Quanlynhansu(hoten_value, quequan_value, maphongban_value, luonghientai_value, ngayvaocongty_value)
    
    # thêm đối tượng vào danh sách
    list_quanlynhansu.append(quanlynhansu)
    showinfo("Quan Ly Nhan Su", "Thong Tin Nhan Su Da Duoc Them Thanh Cong!")
    hoten.delete(0, END)
    quequan.delete(0, END)
    maphongban.delete(0, END)
    luonghientai.delete(0, END)
    ngayvaocongty.delete(0, END)

    
def timkiem_thongtinnhansu():
    timkiem_ten_value = timkiem_ten.get()
    for quanlynhansu in list_quanlynhansu:
        if timkiem_ten_value in quanlynhansu.hoten:
            showinfo("ket qua tim kiem", f"{quanlynhansu.hoten} - QueQuan: {quanlynhansu.quequan} - maphongban: {quanlynhansu.maphongban} -luonghientai: {quanlynhansu.luonghientai} - ngayvaocongty: {quanlynhansu.ngayvaocongty}")
            break
        else:
            showinfo("ket qua tim kiem", "Khong co ten nao nhu vay, cam on")
            
    timkiem_ten.delete(0, END)
    
def capnhat_thongtinnhansu():
    hoten_value = hoten.get()
    quequan_value = quequan.get()
    maphongban_value = maphongban.get()
    luonghientai_value = luonghientai.get()
    ngayvaocongty_value = ngayvaocongty.get()
    
    for quanlynhansu in list_quanlynhansu:
        if hoten_value == quanlynhansu.hoten:
            quanlynhansu.hoten = hoten_value
            quanlynhansu.quequan = quequan_value
            quanlynhansu.maphongban = maphongban_value
            quanlynhansu.luonghientai = luonghientai_value
            quanlynhansu.ngayvaocongty = ngayvaocongty_value
            #showinfo("Cap nhat thanh cong", f"Da cap nhat thanh cong thong tin cho nhan su: {quanlynhansu.hoten}")
    
    hoten.delete(0, END)
    quequan.delete(0, END)
    maphongban.delete(0, END)
    luonghientai.delete(0, END)
    ngayvaocongty.delete(0, END)
    showinfo("Cap nhat thanh cong", f"Da cap nhat thanh cong thong tin cho nhan su: {quanlynhansu.hoten}")
    
def xoa_thongtinnhansu():
    timkiem_ten_value = timkiem_ten.get()
    for i in range(len(list_quanlynhansu)):
        if timkiem_ten_value == list_quanlynhansu[i].hoten:
            del list_quanlynhansu[i]
            showinfo("ket qua tra ve", "da xoa thanh cong nhan su")
            break
        else:
            showinfo("ket qua tim kiem", "Khong co ten nao can xoa nhu vay, cam on")
            break
    timkiem_ten.delete(0, END)
    
    
    
them_btn = Button(window, text="Them", command=them_thongtinnhansu)

timkiem_btn = Button(window, text="Tim Kiem", command=timkiem_thongtinnhansu)
capnhat_btn = Button(window, text="Cap Nhat", command = capnhat_thongtinnhansu)
xoa_btn = Button(window, text="Xoa", command=xoa_thongtinnhansu)

#intro_lbl.grid(row=0, column=0)
intro_lbl.grid(row=0, column=0, columnspan=2)   # columnspan chiếm không gian cột là bao nhiêu
hoten_lbl.grid(row=1, column=0)
quequan_lbl.grid(row=2, column=0)
maphongban_lbl.grid(row=3, column=0)
luonghientai_lbl.grid(row=4, column=0)
ngayvaocongty_lbl.grid(row=5, column=0)

hoten.grid(row=1, column=1)
quequan.grid(row=2, column=1)
maphongban.grid(row=3, column=1)
luonghientai.grid(row=4, column=1)
ngayvaocongty.grid(row=5, column=1)

them_btn.grid(row=6, column=1)
capnhat_btn.grid(row=6, column=2)

timkiem_lbl.grid(row=7, column=0)
timkiem_ten.grid(row=7, column=1)
timkiem_btn.grid(row=7, column=2)
xoa_btn.grid(row=7, column=3)


window.mainloop()
```
